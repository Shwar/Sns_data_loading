# Loading and reading datasets provided by Seaborn


```python
import seaborn as sns
datasets = sns.get_dataset_names()
print(datasets)
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
planets = sns.load_dataset('planets')
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
```

```python
iris
```

```python
sns.scatterplot(x='species', y='sepal_length', data=iris)
```

```python
sns.scatterplot(x='tip', y='total_bill', data=tips, hue='day', size='size', palette ='YlGnBu')
```

```python
sns.histplot(tips['tip'], kde=False, bins=15)
```

```python
sns.displot(tips['tip'], kde=True, bins=30)
```

```python
sns.barplot(x ='sex', y ='tip', data=tips)
```

```python
sns.boxplot(x='day', y='tip',hue='sex', palette='YlGnBu', data=tips)
```

```python
sns.stripplot(x='day', y='tip', data=tips, hue='sex', palette='YlGnBu', dodge=True)
```

```python
sns.jointplot(x='tip', y ='total_bill', data = tips,shade='true', kind='kde', cmap='YlGnBu')
```

```python
sns.jointplot(x='tip', y ='total_bill', data = tips, kind='reg')
```

```python
sns.jointplot(x='tip', y ='total_bill', data = tips, kind='hex', cmap='YlGnBu')
```

```python
import matplotlib.pyplot as plt
sns.pairplot(iris)
plt.show()
```

```python
sns.pairplot(titanic.select_dtypes(['number']), hue='pclass')
```
